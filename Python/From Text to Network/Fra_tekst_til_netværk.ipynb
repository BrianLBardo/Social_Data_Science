```python
import pandas as pd

df = pd.read_excel(r'data/sample.xlsx')

stopord = pd.read_excel(r'data/stopord2.xlsx')
stopordsliste = stopord['word'].tolist()

texts = df.Text

minimums_tekst_hits = 3
```

### Edge table

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(min_df = minimums_tekst_hits, stop_words = stopordsliste)
transformed_documents = vectorizer.fit_transform(texts)

transformed_documents_as_array = transformed_documents.toarray()
count_df = pd.DataFrame(transformed_documents_as_array, columns = vectorizer.get_feature_names_out())

count_df.head(10)
```

```python
# fjern alle kolonner med tal i navnet
count_df = count_df[count_df.columns.drop(list(count_df.filter(regex = '[0-9]')))]
```

```python
# fjern 1-værdierne i diagonalen
from numpy import fill_diagonal

word_cors = count_df.corr()
fill_diagonal(word_cors.values,0)
word_cors.head(10)
```

```python
import numpy as np

word_cors = word_cors.mask(np.tril(np.ones(word_cors.shape)).astype(bool))

word_cors
```

```python
melted_df = pd.melt(word_cors.reset_index(), id_vars = 'index', value_name = "coefficient")

melted_df.head(10)
```

```python
filtered_df = melted_df.nlargest(40000, "coefficient")

filtered_df
```

```python
filtered_df = filtered_df.rename(columns={"index": "Source", "variable": "Target"})

filtered_df.to_csv("edge_table.csv", index = False)
```

### Node table

```python
node_table = count_df.transpose()

node_table["sum"] = node_table.sum(axis = 1)

node_table = node_table["sum"]

node_table = pd.DataFrame(node_table)

node_table.reset_index(inplace=True)

node_table = node_table.rename(columns={"index": "Id"})

node_table.to_csv("node_table.csv", index = False)
```

### Opdateret edge table

```python
import re
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    return text

count_model = CountVectorizer(min_df = minimums_tekst_hits, preprocessor = preprocess_text, ngram_range = (1, 1), stop_words = stopordsliste)
X = count_model.fit_transform(texts)
X[X > 0] = 1
Xc = (X.T * X)
Xc.setdiag(0)

X_as_array = Xc.toarray()
X_df = pd.DataFrame(X_as_array,
                    columns = count_model.get_feature_names_out(),
                    index = count_model.get_feature_names_out())

X_df.head(10)
```

```python
melted_X = pd.melt(X_df.reset_index(), id_vars = 'index', value_name = "count")

melted_X = melted_X.rename(columns={"index": "Source", "variable": "Target"})

melted_X.head(10)
```

```python
edge_table_update = pd.merge(filtered_df, melted_X,  how ='left',
                             left_on = ["Source", "Target"],
                             right_on = ["Source", "Target"])

edge_table_update
```

```python
edge_table_update["synthetic"] = edge_table_update["coefficient"] * edge_table_update["count"]

edge_table_update
```

```python
edge_table_update.to_csv("edge_table.csv", index = False)
```
