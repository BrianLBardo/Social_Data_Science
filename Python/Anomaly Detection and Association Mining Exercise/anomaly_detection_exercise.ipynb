In this notebook, we will try to find if Google search trends data on the keyword "shooting" contains any anomalous value(s). The below code gets Google trends data using the pytrends library. Your task is to develop a strategy to detect anomalous data points, if any exists.

```python
import pandas as pd
from statsmodels.tsa.seasonal import STL
import matplotlib.pyplot as plt
from datetime import datetime
from pytrends.request import TrendReq

pytrend = TrendReq(hl='en-US', tz=360)
pytrend.build_payload(kw_list=['shooting'], timeframe= '2016-01-01 2022-12-31')
gtrends_data = pytrend.interest_over_time()

plt.figure(figsize=(15,5))
plt.plot(gtrends_data.shooting)
plt.title('Shooting - Search Interest', fontsize=20)
```

```python
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
from statsmodels.tsa.seasonal import STL
res = STL(gtrends_data.shooting).fit()

res.plot()
plt.show()
```
